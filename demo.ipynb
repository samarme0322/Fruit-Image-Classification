```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

model = load_model('models/vgg2.h5')

class_labels = ['Apple 5', 'Apricot 1', 'Avocado 1', 'Banana 3', 'Beans 1','Blackberrie 1', 'Cabbage white 1', 'Cactus fruit 1', 'Carrot 1', 'Corn 1']
```

```python
cap = cv2.VideoCapture(0)

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 800)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 600)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    img = cv2.resize(frame, (100, 100))
    img_array = img.astype('float32') / 255.0 
    img_array = np.expand_dims(img_array, axis=0) 

    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions[0])
    confidence = predictions[0][predicted_class]

    label = f"{class_labels[predicted_class]}: {confidence*100:.2f}%"

    cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX,
                1, (0, 255, 0), 2, cv2.LINE_AA)

    cv2.imshow('class', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
